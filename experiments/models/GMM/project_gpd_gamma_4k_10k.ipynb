```python
""" load training data """

FILE_NAME = 'traindata_gamma.npz'

npzfile = np.load(path + FILE_NAME)
train_data = npzfile['arr_0']
meta_info = npzfile['arr_1']
batch_size = meta_info[0]
ndim_x = meta_info[1]

print('training data loaded from .npz file. Rows: %d ' % len(train_data), ' Columns: %d ' % len(train_data[0]), 'Batch_size: %d' % batch_size, 'ndim_x: %d' % ndim_x)
```

```python
""" create models and train them """

n_epoch = 10000
FILE_NAME = 'est_models.pkl'

Y = train_data[:,0]
X = train_data[:,1:]

emm_model = GPDExtremeValueMixtureDensityNetwork("EMM_GPD", ndim_x=ndim_x, n_centers=10, ndim_y=1, n_training_epochs=n_epoch, hidden_sizes=(32, 32),verbose_step=math.floor(n_epoch/10),learning_rate=1e-3)
emm_model.fit(X, Y)

gmm_model = MixtureDensityNetwork("GMM", ndim_x=ndim_x, n_centers=10, ndim_y=1,n_training_epochs=n_epoch,hidden_sizes=(32, 32))
gmm_model.fit(X, Y)

pmm_model = ExtremeValueMixtureDensityNetwork("EMM_PL", ndim_x=ndim_x, n_centers=10, ndim_y=1, n_training_epochs=n_epoch, hidden_sizes=(32, 32),verbose_step=math.floor(n_epoch/10),learning_rate=1e-3)
pmm_model.fit(X, Y)

models = [emm_model,gmm_model,pmm_model]
with open(path + FILE_NAME, 'wb') as output:
    pickle.dump(models, output, pickle.HIGHEST_PROTOCOL)
```

```python
""" load trained models """

FILE_NAME = 'est_models.pkl'

with open(path+FILE_NAME, 'rb') as input:
    sess = tf.Session()
    with sess:
        d_models = pickle.load(input)

emm_model = d_models[0]
emm_model._setup_inference_and_initialize()

gmm_model = d_models[1]
gmm_model._setup_inference_and_initialize()

pmm_model = d_models[2]
pmm_model._setup_inference_and_initialize()

n_epoch = emm_model.n_training_epochs

print(d_models)
print('n_epoch: %d'%n_epoch)
```
